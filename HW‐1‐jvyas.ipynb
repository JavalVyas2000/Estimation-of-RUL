# **Problem 1 (a)**

We try and minimize the verticle distance between points and ground truth and the predicted values, we do this by using two parameters, i.e. $\theta_0$ and $\theta_1$. Thus, the equation is just the distance formula which will act as the cost funtion. The equation is as follows:

\begin{align}
  J(\theta) = \Sigma_{i=1}^m (y^i - h_\theta(x)^i)^2 \\
  where,\ h_\theta(x)^i =  \theta_0 + \theta_1x \\
\end{align}

Thus, to minimize the cost funtion we differentiate it w.r.t. $\theta$ and equate it to 0.

\begin{align}
  \frac{\partial J(\theta)}{\partial \theta} = \Sigma_{i=1}^m \frac{\partial  (y^i - h_\theta(x)^i)^2}{\partial \theta} \\
  \\
\end{align}

\begin{align}
  h_\theta(x) = \theta_0 + \theta_1x  \\
  \\
\end{align}

Solving w.r.t. $\theta_0$
\begin{align}
  \frac{\partial J(\theta)}{\partial \theta_0} = \Sigma_{i=1}^m \frac{\partial  (y^i - (\theta_0 + \theta_1x)^i)^2}{\partial \theta_0}\\
  = -\Sigma_{i=1}^m 2(y^i -(\theta_0 + \theta_1x^i))\\
  \\
  \end{align}

  $\therefore \Sigma_{i=1}^m 2(y^i -(\theta_0 + \theta_1x^i)) = 0\\
  \therefore \Sigma_{i=1}^m (y^i -(\theta_0 + \theta_1x^i))=0\\
  \therefore \Sigma_{i=1}^m (y^i -\theta_1x^i)) = \Sigma_{i=1}^m\theta_0\\
  \therefore \Sigma_{i=1}^m (y^i -\theta_1x^i)) = m\theta_0\\
  \therefore \theta_0 =  \frac{\Sigma_{i=1}^m (y^i -\theta_1x^i)}{m}\\
  \\
  $

Putting in the values for the y and x and getting equation in x and y

  $\theta_0 =  \frac{\Sigma_{i=1}^m (y^i -\theta_1x^i)}{m}\\
  \theta_0 = \frac{2-\theta_1(-2) + 4 - \theta_1(2) + 8 - \theta_1(3) + 11 - \theta_1(5) + 17-\theta_1(4)}{5}\\
  \\
  5\theta_0 = 42 -12\theta_1\\
  5\theta_0 -12\theta_1=42-----(1)\\
  $

Solving w.r.t. $\theta_1$

\begin{align}
 \frac{\partial J(\theta)}{\partial \theta_1} = \Sigma_{i=1}^m \frac{\partial  (y^i - (\theta_0 + \theta_1x)^i)^2}{\partial \theta_1}\\
  = -\Sigma_{i=1}^m 2x^i(y^i -(\theta_0 + \theta_1x^i))\\
  \\
  \end{align}

  $\therefore \Sigma_{i=1}^m 2 x^i (y^i -(\theta_0 + \theta_1x^i)) = 0\\
  \therefore \Sigma_{i=1}^m x^i(y^i -(\theta_0 + \theta_1x^i))=0\\
  \therefore \Sigma_{i=1}^m x^iy^i - \theta_0x^i - \theta_1(x^i)^2 = 0\\
  \therefore \Sigma_{i=1}^m (x^iy^i - \theta_0x^i)) = \theta_1(x^i)^2\\
  \therefore \theta_1 =  \frac{\Sigma_{i=1}^m(x^iy^i - \theta_0x^i)}{\Sigma_{i=1}^m (x^i)^2}\\
  \\
$

Solving for $\theta_0$ and $\theta_1$ simultaneously, we get,

```python
import numpy as np
import matplotlib.pyplot as plt

def linear_reg(X,y):
	theta=np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
	return theta
x=np.array([-2,2,3,5,4])
y=np.array([2,4,8,11,17])
X=[]
for i in range(len(x)):
	X.append([1,x[i]])
X=np.array(X)
y=np.array(y)
#Normal Equation: theta = (X X^T)-1.(X^T Y)
theta=linear_reg(X,y)

print(f'Theta_0 = {theta[0]} and Theta_1 = {theta[1]}')
# print(theta)
```

The answers in both the cases is the same, it doesn't defer till the 4th decimal which is the best precision that I could do by hand. **Both answers match.**

### **Derive the Normal equation that you used to solve the above problem. Also, explain Gradient Descent and describe three types of Gradient Descent.**

The cost function is given by :
\begin{align}
  J(\theta) = \Sigma_{i=1}^m (y^i - h_\theta(x)^i)^2 \\
  where,\ h_\theta(x)^i =  \theta_0 + \theta_1x \\
\end{align}

Thus, to minimize the cost funtion we differentiate it w.r.t. $\theta$ and equate it to 0. For similicity we will consider $\theta$ as a matrix of size (2,1) and x to be a matrix of size(n,2) where n is the length of the dataset. So, now the first coloumn of x is all filled with 1 as $\theta_0$ is multiplied with 1, and $\theta_1$ is multiplied with the data input given by the user.

Thus,  

\begin{align}
  \frac{d(J(\theta))}{d(\theta)} = \Sigma_{i=1}^m \frac{d(y^i - h_\theta(x)^i)^2}{d\theta} \\
  \\
\end{align}

\begin{align}
  h_\theta(x) = \theta_0 + \theta_1x  \\
  \\
\end{align}

\begin{align}
  \frac{d(J(\theta))}{d(\theta_0)} = \Sigma_{i=1}^m \frac{2x(y - \theta x))}{\partial \theta_0}\\
  \\
  = -\Sigma_{i=1}^m 2x(y -\theta x))\\
  \\
  \end{align}

The update rule of for theta is given by, 

$\theta^{ t+1}$ = $\theta^{ t}$ -  $\frac{d(J(\theta))}{d(\theta)}$

$\theta^{ t+1}$ = $\theta^{ t}$ - $\frac{1}{m}$ $\Sigma_{i=1}^m 2x(y -\theta x))$

Where  is the learning rate which determines the step size of the function.

There are 3 types of Gradient Descend, namely:

if m=1 : Stocastic Gradient Descend 

if m=ma : Partial Gradient Descend
